# Chi Square Test (Association)

```python
import pandas as pd
from scipy.stats import chi2_contingency

file_path = 'dataset_final.csv'
df = pd.read_csv(file_path)

contingency_table = pd.crosstab(df['Are you a veteran?'], df['Are you currently experiencing homelessness?'])

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")


alpha = 0.05
if p < alpha:
    print("There is a significant association between veteran status and homelessness.")
else:
    print("There is no significant association between veteran status and homelessness.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency


file_path = 'dataset_final.csv'
df = pd.read_csv(file_path)

contingency_table = pd.crosstab(df['Are you a veteran?'], df['Are you currently experiencing homelessness?'])


sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt='d', cbar=False)
plt.title("Contingency Table - Veteran Status vs. Homelessness")
plt.xlabel("Homelessness")
plt.ylabel("Veteran Status")
plt.show()


chi2, p, _, _ = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

alpha = 0.05
if p < alpha:
    print("There is a significant association between veteran status and homelessness.")
else:
    print("There is no significant association between veteran status and homelessness.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

file_path = 'dataset_final.csv' 
df = pd.read_csv(file_path)


time_points = df['What is your current marital status?'].unique()

for time_point in time_points:
    subset_df = df[df['What is your current marital status?'] == time_point]
    contingency_table = pd.crosstab(subset_df['Are you a veteran?'], subset_df['Are you currently experiencing homelessness?'])

   
   
    sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt='d', cbar=False)
    plt.title(f"Contingency Table - Veteran Status vs. Homelessness (Point:: {time_point})")
    plt.xlabel("Homelessness")
    plt.ylabel("Veteran Status")
    plt.show()
```
